```python
all_df['Отклонение сайт\сайт'] = all_df['Отклонение сайт\сайт'] * (-1)
```

```python
all_df['Отклонение МРЦ/сайт'] = all_df['Отклонение МРЦ/сайт'] * (-1)
```

```python
d = {'True' : False, 'False' : True}
all_df['Демпинг сайт\сайт'] = all_df['Отклонение сайт\сайт'].replace('', 0).astype('float64') < 0
```

```python
all_df['Демпинг МРЦ'] = all_df['Отклонение МРЦ/сайт'].replace('', 0).astype('float64') < 0
```

```python
all_df['Отклонение МРЦ/сайт'] = all_df['Отклонение МРЦ/сайт'].replace('', 'н/д')
```

```python
all_df['Отклонение сайт\сайт'] = all_df['Отклонение сайт\сайт'].replace('', 'н/д')
```

```python
all_df.loc[all_df['Отклонение МРЦ/сайт']== 'н/д',  'Демпинг МРЦ'] = 'н/д' 
```

```python
all_df.loc[all_df['Отклонение сайт\сайт']== 'н/д',  'Демпинг сайт\сайт'] = 'н/д' 
```

```python
with pd.ExcelWriter('output6.xlsx', options={'strings_to_urls': False}) as writer:
    mrc_df.to_excel(writer, index=False, sheet_name='Отклонения по МРЦ')
    all_df.to_excel(writer, index=False, sheet_name='Все данные')
```
